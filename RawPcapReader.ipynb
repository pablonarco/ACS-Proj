Importing layers to open pkts

```
from scapy.layers.l2 import Ether
from scapy.layers.dns import DNS, DNSQR
from scapy.layers.inet import IP, TCP, UDP
```

Opening ethernet header, note pkt no 9 having type 2054 which is arp pkt

```
i = 0
for (pkt_data, pkt_metadata) in RawPcapReader('/content/drive/My Drive/Colab Notebooks/onelakhbot.pcap'):
    type = Ether(pkt_data)
    print (i,type.fields, type.type)
    i+=1
```

0pening IP header, note pkts with protocol 17 which is a DNS pkt

```
i = 0
for (pkt_data, pkt_metadata) in RawPcapReader('/content/drive/My Drive/Colab Notebooks/test.pcap'):
    pkt2 = Ether(pkt_data)
    if pkt2.type == 2048:
        type3 = pkt2[IP]
        print (i,type3.fields)
    i+=1
```

Pkt no, time and DNS query

```
i = 0
for (pkt_data, pkt_metadata) in RawPcapReader('/content/drive/My Drive/Colab Notebooks/test.pcap'):
    pkt2 = Ether(pkt_data)
    if pkt2.type == 2048:
        pkt3 = pkt2[IP]
        if pkt3.proto == 17:
            time = str(pkt_metadata.sec) +'.' + str(pkt_metadata.usec)
            print (i, time, pkt3[DNS][DNSQR].qname)
    i+=1
```

We have assumed all UDP pkts as DNS till not and it worked in the sample file with 141 pkts only. Issues faced in file with 100000 pkts. Code below to identify the problem. Few UDP pkts with different protocols identified.

```
pca = rdpcap ('/content/drive/My Drive/Colab Notebooks/onelakhbot.pcap')
i = 0
for a in pca:
    if ('UDP' in a) & ('DNS' not in a):
        print ('UDP but not DNS pkt no:', i)
    i += 1
```

Solving the above problem and checking each UDP pkt for DNS headers and then printing

```
i = 0
for (pkt_data, pkt_metadata) in RawPcapReader('/content/drive/My Drive/Colab Notebooks/onelakhbot.pcap'):
    pkt2 = Ether(pkt_data)
    i +=1
    if pkt2.type == 2048:
        type3 = pkt2[IP]
        if 'UDP' in type3:
            if 'DNS' in type3:
                if (type3.proto == 17) :
                    time = str(pkt_metadata.sec) +'.' + str(pkt_metadata.usec)
                    query = type3[UDP]
                    print (i, time, type3[UDP][DNSQR].qname)
```
